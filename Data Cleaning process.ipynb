# Missing Value Input bt without delete the dataset 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"F:\ML_dataset\set_2\train.csv")
```

```python
pd.set_option('display.max_columns',None) #to show the all row if all rows are not visible 
pd.set_option('display.max_columns',None)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.tail(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
missing_value_per =df.isnull().sum()/df.shape[0]*100
missing_value_per
```

```python
percent_missing = df.isnull().sum() * 100 / len(df) # or this formula to find out the missing value 
percent_missing 
```

```python
missing_value_clm_gre_10 = missing_value_per[missing_value_per>10].keys()
missing_value_clm_gre_10
```

```python
df2_drop_clm = df.drop(columns =missing_value_clm_gre_10)
df2_drop_clm.shape
```

```python
df3_num =df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head(6)
```

```python
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]# this function shows that in which row the null value is present
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(3,2,i+1)
    sns.distplot(df3_num[var],bins=10,kde_kws={'linewidth':5,'color':'#DC143C'})
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean()) #
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(3,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':5,'color':'red'},label='Original')
    sns.distplot(df4_num_mean[var],bins=10,kde_kws={'linewidth':3,'color':'green'},label='Mean')
    plt.legend()
```

```python
df5_num_median = df3_num.fillna(df3_num.median()) #
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(3,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':5,'color':'red'},label='Original')
    sns.distplot(df4_num_mean[var],bins=10,kde_kws={'linewidth':3,'color':'green'},label='Mean')
    sns.distplot(df5_num_median[var],bins=10,kde_kws={'linewidth':3,'color':'blue'},label='Median')
    plt.legend()
```

```python
#for cheaking the outlair
#the impact of outlair in ML algorithem is it effects the accuracy 
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
    plt.legend()
```

```python
df_concat= pd.concat([df3_num[missing_num_var],df4_num_mean[missing_num_var],df5_num_median[missing_num_var]],axis=1)
df_concat[df_concat.isnull().any(axis=1)]
                                    
```
