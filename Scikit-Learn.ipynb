```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
train=pd.read_csv(r"F:\ML_dataset\set_2\train.csv")
test=pd.read_csv(r"F:\ML_dataset\set_2\test.csv")
print(train.shape)
print(test.shape)
```

```python
X_train=train.drop(columns="TARGET(PRICE_IN_LACS)")
y_train=train["TARGET(PRICE_IN_LACS)"]
X_test = test.copy()
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
```

# missing value imputation

```python
isnull_var = X_train.isnull().sum()
isnull_var
```

```python
miss_val=X_train.select_dtypes(include=["int64","float64"]).columns
num_val_miss = [var for var in miss_val if isnull_var[var]>0]
num_val_miss
```

```python
cat_val=X_train.select_dtypes(include=["O"]).columns
cat_val_miss=[var for var in cat_val if isnull_var[var]>0]
cat_val_miss
```

```python
#mode value -- most frequent value
```

```python
num_var_mean = ['BHK_NO.']
num_var_median = ['SQUARE_FT','LONGITUDE']
cat_var_mode = ['ADDRESS']
#num_ver_missing = ['UNDER_CONSTRUCTION','RERA','READY_TO_MOVE',]
```

```python
num_var_mean_imputer = Pipeline(steps=[("inputer",SimpleImputer(strategy = "mean"))])
num_var_median_imputer = Pipeline(steps=[("inputer",SimpleImputer(strategy = "median"))])
cat_var_most_freq_imputer = Pipeline(steps=[("inputer",SimpleImputer(strategy = "most_frequent"))])
#num_var_missing_imputer = Pipeline(steps=[("inputer",SimpleImputer(strategy = "constant",fill_value = "missing"))])
```

```python
prepocessor = ColumnTransformer (transformers=[("mean_imputer",num_var_mean_imputer,num_var_mean),("median_imputer",num_var_median_imputer,num_var_median),("mode_imputer",cat_var_most_freq_imputer,cat_var_mode)])
#,("missing_imputer",num_var_missing_imputer,num_ver_missing)
```

```python
prepocessor.fit(X_train)
```

```python
prepocessor.named_transformers_["mean_imputer"].named_steps["inputer"].statistics_
```

```python
train['BHK_NO.'].mean()
```

```python
X_train_clear= prepocessor.transform(X_train)
X_test_clear= prepocessor.transform(X_test)
```
